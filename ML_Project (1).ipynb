```python
counts = df1["smoking_history"].value_counts()

smoker_categories = ["former", "current", "not current", "ever"]
non_smoker_categories = ["never"]


smokers = sum(counts.get(category, 0) for category in smoker_categories)
non_smokers = sum(counts.get(category, 0) for category in non_smoker_categories)
no_info = counts.get("No Info", 0)


total = smokers + non_smokers


smoker_proportion = smokers / total
non_smoker_proportion = non_smokers / total


smokers_from_no_info = int(no_info * smoker_proportion)
non_smokers_from_no_info = no_info - smokers_from_no_info

no_info_indices = df1[df1["smoking_history"] == "No Info"].index
np.random.seed(42)
smoker_indices = np.random.choice(no_info_indices, size=smokers_from_no_info, replace=False)
non_smoker_indices = list(set(no_info_indices) - set(smoker_indices))

df1.loc[smoker_indices, "smoking_history"] = "Smoker"
df1.loc[non_smoker_indices, "smoking_history"] = "Non-Smoker"


df1["smoking_history"] = df1["smoking_history"].replace(smoker_categories, "Smoker")
df1["smoking_history"] = df1["smoking_history"].replace(non_smoker_categories, "Non-Smoker")

print(df1["smoking_history"].value_counts())
```

```python
X1 = df.drop('cardio', axis=1)
y1 = df['cardio']

smote = SMOTE(random_state=42)
X1_resampled, y1_resampled = smote.fit_resample(X1, y1)
X_train1, X_test1, y_train1, y_test1 = train_test_split(X1_resampled, y1_resampled, test_size=0.2, random_state=42)



```

```python
from sklearn.feature_selection import SelectKBest, f_classif

selector = SelectKBest(f_classif, k=10)
X_train1 = selector.fit_transform(X_train1, y_train1)
X_test1 = selector.transform(X_test1)
```

```python

selected_mask = selector.get_support()


feature_names = X1.columns


selected_features = feature_names[selected_mask]
removed_features = feature_names[~selected_mask]

print("Selected Features:", list(selected_features))
print("Removed Features:", list(removed_features))
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, roc_auc_score
from sklearn.model_selection import KFold, GridSearchCV

model1 = LogisticRegression(max_iter=3000)

k = 5
kf = KFold(n_splits=k, shuffle=True, random_state=42)


param_grid = {
    'C': [0.01, 0.1, 0.5, 1, 10, 100],
    'solver': ['liblinear', 'lbfgs'],
    'penalty': ['l1', 'l2']
}


grid_search = GridSearchCV(estimator=model1, param_grid=param_grid,
                           cv=kf, scoring='roc_auc', n_jobs=-1, verbose=1)

grid_search.fit(X_train1, y_train1)


best_params = grid_search.best_params_
print(f"Best hyperparameters found: {best_params}")


best_model = grid_search.best_estimator_


accuracy_list = []
roc_auc_list = []

```

```python
for fold, (train_index, val_index) in enumerate(kf.split(X_train1)):
    print(f"Fold {fold + 1}/{k}")


    X_train_fold, X_val_fold = X_train1[train_index], X_train1[val_index]
    y_train_fold, y_val_fold = y_train1.iloc[train_index], y_train1.iloc[val_index]


    best_model.fit(X_train_fold, y_train_fold)


    y_pred_fold = best_model.predict(X_val_fold)
    y_pred_proba_fold = best_model.predict_proba(X_val_fold)[:, 1]


    fold_accuracy = accuracy_score(y_val_fold, y_pred_fold)
    fold_roc_auc = roc_auc_score(y_val_fold, y_pred_proba_fold)


    accuracy_list.append(fold_accuracy)
    roc_auc_list.append(fold_roc_auc)

    print(f"Fold {fold + 1} Accuracy: {fold_accuracy:.2f}, ROC AUC: {fold_roc_auc:.2f}")


average_accuracy = np.mean(accuracy_list)
average_roc_auc = np.mean(roc_auc_list)

print("\nCross-Validation Results")
print(f"Average Accuracy: {average_accuracy:.2f}")
print(f"Average ROC AUC: {average_roc_auc:.2f}")


best_model.fit(X_train1, y_train1)


y_pred_test = best_model.predict(X_test1)
y_pred_proba_test = best_model.predict_proba(X_test1)[:, 1]

test_accuracy = accuracy_score(y_test1, y_pred_test)
test_roc_auc = roc_auc_score(y_test1, y_pred_proba_test)

print("\nTest Set Results")
print(f"Test Accuracy: {test_accuracy:.2f}")
print(f"Test ROC AUC: {test_roc_auc:.2f}")
```
